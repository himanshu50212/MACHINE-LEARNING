```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('weatherAUS.csv')
```

```python
dataset
```

```python
dataset.drop(dataset.tail(1).index,inplace=True)
```

```python
dataset = dataset.set_index('Location')
```

```python
dataset = dataset.drop(['Date'], axis=1)
```

```python
print(dataset.shape)
```

```python
dataset.info()
```

```python
dataset = dataset.fillna(dataset.mean())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['RainToday']= le.fit_transform(dataset['RainToday'])
dataset['RainTomorrow']= le.fit_transform(dataset['RainTomorrow'])
```

```python
dataset
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(data= dataset.corr(), cmap="YlGnBu", annot= True)
plt.title("Pairwise correlation of all the columns in the dataframe ")
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
dataset.WindGustDir = OneHotEncoder().fit_transform(dataset[["WindGustDir"]]).toarray()
```

```python
dataset.WindDir9am = OneHotEncoder().fit_transform(dataset[["WindDir9am"]]).toarray()
```

```python
dataset.WindDir3pm = OneHotEncoder().fit_transform(dataset[["WindDir3pm"]]).toarray()
```

```python
dataset
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lclassifier = LogisticRegression(random_state = 0)
lclassifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
ly_pred = lclassifier.predict(X_test)
```

```python
print(accuracy_score(y_test, ly_pred))
```
